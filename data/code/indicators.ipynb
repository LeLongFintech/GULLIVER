```python
import pandas as pd
import numpy as np
```

```python
ohlcv = pd.read_csv(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\data\data_cleaned\OHLCV_Merge.csv")
ohlcv
```

```python
ohlcv['date'] = pd.to_datetime(ohlcv['date'], errors='coerce')
ohlcv = ohlcv.dropna(subset=['date'])

ohlcv['ticker'] = ohlcv['ticker'].astype(str).str.strip().str.upper()

target_years = [2020, 2021, 2022, 2023, 2024]
ohlcv['_year'] = ohlcv['date'].dt.year
ohlcv_y = ohlcv.loc[ohlcv['_year'].isin(target_years)].copy()

idx = ohlcv_y.groupby(['ticker', '_year'])['date'].idxmax()

ohlcv = (
    ohlcv_y.loc[idx]
    .rename(columns={'_year': 'year'})
    .sort_values(['ticker', 'date'])
    .reset_index(drop=True)
)
ohlcv
```

```python
balance_sheet = pd.read_csv(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\data\data_cleaned\Balance_sheet.csv")
balance_sheet
```

```python
cash_flow = pd.read_csv(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\data\data_cleaned\Cash_flow.csv")
cash_flow
```

```python
income_statement = pd.read_csv(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\data\data_cleaned\Income_statement.csv")
income_statement
```

```python
indicators = pd.read_excel(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\data\raw_data\info.xlsx")
indicators
```

```python
shares_outstanding = indicators.copy()
eps    = income_statement[' Lãi cơ bản trên cổ phiếu']
profit = income_statement[' Cổ đông của Công ty mẹ']

shares_outstanding['shares_outstanding'] = 0.0
mask = eps.notna() & (eps != 0)

shares_outstanding.loc[mask, 'shares_outstanding'] = (profit.loc[mask] / eps.loc[mask])

shares_outstanding['shares_outstanding'] = np.rint(shares_outstanding['shares_outstanding']).astype('Int64')
shares_outstanding
```

```python
ohlcv['ticker'] = ohlcv['ticker'].astype(str).str.strip().str.upper()
ohlcv['year']   = pd.to_numeric(ohlcv['year'], errors='coerce').astype('Int64')

indicators['Mã'] = indicators['Mã'].astype(str).str.strip().str.upper()
indicators['Năm'] = pd.to_numeric(indicators['Năm'], errors='coerce').astype('Int64')

indicators['P/E'] = np.nan

for i in range(len(indicators)):
    tkr = indicators.loc[i, 'Mã']
    year = indicators.loc[i, 'Năm']
    eps = income_statement.loc[i, ' Lãi cơ bản trên cổ phiếu']  # giữ cùng thứ tự
    
    # Tìm giá close khớp mã và năm
    match = ohlcv[(ohlcv['ticker'] == tkr) & (ohlcv['year'] == year)]
    
    if not match.empty and pd.notna(eps) and eps != 0:
        close_val = match.iloc[-1]['close']
        indicators.loc[i, 'P/E'] = round(close_val * 1000 / eps, 2)
    else:
        indicators.loc[i, 'P/E'] = np.nan

indicators['P/B'] = np.nan
for i in range(len(indicators)):
    tkr = indicators.loc[i, 'Mã']
    year = indicators.loc[i, 'Năm']
    vcsh = balance_sheet.loc[i, " VỐN CHỦ SỞ HỮU"] 
    shares = shares_outstanding.loc[i, "shares_outstanding"]
    
    # Tìm giá close khớp mã và năm
    match = ohlcv[(ohlcv['ticker'] == tkr) & (ohlcv['year'] == year)]
    
    if not match.empty and pd.notna(shares) and shares != 0:
        close_val = match.iloc[-1]['close']
        indicators.loc[i, 'P/B'] = round(close_val * 1000 / (vcsh / shares), 2)
    else:
        indicators.loc[i, 'P/B'] = np.nan
```

```python
indicators['EPS'] = round(income_statement[' Lãi cơ bản trên cổ phiếu'], 2)
indicators['BVPS'] = round(balance_sheet[' VỐN CHỦ SỞ HỮU'] / shares_outstanding['shares_outstanding'])
indicators['ROE (%)'] = round(income_statement[' Cổ đông của Công ty mẹ'] / balance_sheet[' VỐN CHỦ SỞ HỮU'] * 100, 2)
indicators['ROA (%)'] = round(income_statement[' Cổ đông của Công ty mẹ'] / balance_sheet[' TỔNG CỘNG TÀI SẢN'] * 100, 2)
indicators['Nợ / Vốn chủ sở hữu (DE)'] = round(balance_sheet[' NỢ PHẢI TRẢ'] / balance_sheet[' VỐN CHỦ SỞ HỮU'], 2)
indicators['Biên lợi nhuận gộp'] = round(income_statement[' Lợi nhuận gộp về bán hàng và cung cấp dịch vụ'] / income_statement[' Doanh thu thuần'], 2)
indicators['Biên lợi nhuận ròng'] = round(income_statement[' Lợi nhuận thuần từ hoạt động kinh doanh'] / income_statement[' Doanh thu thuần'], 2)
indicators['Thanh khoản hiện hành'] = round(balance_sheet[' TÀI SẢN NGẮN HẠN'] / balance_sheet[' Nợ ngắn hạn'], 2)
indicators['Thanh khoản nhanh'] = round((balance_sheet[' TÀI SẢN NGẮN HẠN'] - balance_sheet[' Hàng tồn kho, ròng']) / balance_sheet[' Nợ ngắn hạn'], 2)

indicators
```

```python
import pandas as pd
import numpy as np
```

```python
indicators = pd.read_csv(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\front_end\public\Indicators.csv")
info = pd.read_csv(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\front_end\public\Stock_info.csv")
```

```python
info_sector = info[['Symbol', 'Sector']]

# merge: giữ nguyên tất cả dòng của indicators
merged = indicators.merge(
    info_sector,
    how='left',
    left_on='Mã',
    right_on='Symbol'
)

# nếu không muốn giữ cột Symbol sau merge thì drop đi:
merged = merged.drop(columns=['Symbol'])
```

```python
indicators = indicators.merge(info[['Symbol', 'Sector']],
                              how='left',
                              left_on='Mã',
                              right_on='Symbol') \
                       .drop(columns=['Symbol'])
```

```python
indicators
```

```python
num_cols = merged.select_dtypes(include='number').columns.tolist()

if 'Năm' in num_cols:
    num_cols.remove('Năm')

df = (
    merged
    .groupby(['Sector', 'Năm'])[num_cols]
    .mean()
    .reset_index()
    .sort_values(['Sector', 'Năm'])
    .reset_index(drop=True)
)
df = df.round(2)
df = df.drop(df.columns[[2, 3]], axis=1)
df
```

```python
import pandas as pd
import numpy as np
```

```python
df_HNX = pd.read_csv(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\data\raw_data\CafeF.HNX.Upto02.11.2025.csv")
df_HOSE = pd.read_csv(r"D:\Tài Liệu Học\Học kỳ chính quy\Kỳ 6\Gói phần mềm ứng dụng 1\MID_TERM_PROJECT\data\raw_data\CafeF.HSX.Upto02.11.2025.csv")
```

```python
df_HNX['exchange'] = "HNX"
df_HNX = df_HNX.rename(columns = {
    "<Ticker>": "ticker",
    "<DTYYYYMMDD>": "date",
    "<Open>": "open",
    "<High>": "high",
    "<Low>": "low",
    "<Close>": "close",
    "<Volume>": "volume"
})
```

```python
df_HNX
```

```python
df_HOSE['exchange'] = "HOSE"
df_HOSE = df_HOSE.rename(columns = {
    "<Ticker>": "ticker",
    "<DTYYYYMMDD>": "date",
    "<Open>": "open",
    "<High>": "high",
    "<Low>": "low",
    "<Open>": "open",
    "<Close>": "close",
    "<Volume>": "volume"
})
```

```python
df_HOSE
```

```python
df = pd.concat([df_HNX, df_HOSE], ignore_index=True)
df = df.sort_values(by=["ticker", "date"], ascending=[True, True]).reset_index(drop=True)
df
```

```python
df[["ticker", "exchange"]] = df[["ticker", "exchange"]].astype(str)
df["date"] = pd.to_datetime(df["date"], format = "%Y%m%d")

num_cols = ["open", "high", "low", "close"]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")

df["volume"] = pd.to_numeric(df["volume"], errors="coerce").astype("Int64")

df
```

```python
df = df[(df["date"].dt.year >= 2020) & (df["date"].dt.year <= 2024)].reset_index(drop=True)
df
```
